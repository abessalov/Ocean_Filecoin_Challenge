### 0. Read dataset

```python
df = pd.read_csv('data/FilecoinGreenEnergyNew.csv')

df['Date'] = pd.to_datetime(df.Date)
df['year']  = df.Date.map(lambda x: str(x)[:4])
df['quarter'] = df.year.astype(str) + 'Q' + df.Date.dt.quarter.astype(str).values
df['month'] = df.Date.map(lambda x: str(x)[:7])

df.set_index('Date', inplace = True)
print(df.shape)
df[:3]
```

##### Monthly changes of estimated values

```python
feat = 'Energy_intensity_estimate_bound'
```

```python
df[feat][:3]
```

```python
# monthly differences
df0 = df.groupby('month')[[feat]].last()
df0['diff_abs'] = -df0.diff()
df0['diff_rel'] = -df0.iloc[:,0].pct_change()
df0['color'] = df0['diff_abs'].map(lambda x: 'green' if x > 0 else 'red')
df0.T
```

```python
# last 2 years:
a1, a2, a3 = 39.00, 9.89, 6.07
print(a1,a3,a3-a1)
print(a1,a2,a2-a1)
print(a2,a3,a3-a2)
```

```python
# plotting
_, ax = plt.subplots(1,2, figsize = (18,5))
df1 = df0[4:]
_ = df1.diff_abs.plot(kind = 'bar', ax = ax[0], title = 'Absolute changes', xlabel = '', color = df1.color)
_ = df1.diff_rel.plot(kind = 'bar', ax = ax[1], title = 'Percentage changes', xlabel = '', color = df1.color)
```
